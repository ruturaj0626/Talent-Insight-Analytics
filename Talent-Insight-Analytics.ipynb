```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('/kaggle/input/hr-employee-attrition/HR-Employee-Attrition.csv')
dataset
```

```python
#Data cleaning
dataset.isnull().sum()
```

```python
dataset.dropna(inplace = True)
```

```python
dataset.isnull().sum()
```

```python
businesstravel = dataset['BusinessTravel'].value_counts()
businesstravelDF = pd.DataFrame(businesstravel)
businesstravelDF
```

```python
businesstravel = dataset['BusinessTravel'].value_counts()

# Create the bar plot
plt.figure(figsize=(6, 5))
businesstravel.plot(kind='bar', color=['skyblue','purple','coral'])
plt.xlabel('Business Travel Frequency')
plt.ylabel('Count')
plt.title('Distribution of Business Travel')
plt.xticks(rotation=0)

# Add labels to the bars
for i, count in enumerate(businesstravel):
    plt.text(i, count + 10, str(count), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
#Descriptive Information:
#1) What is the range of ages in the dataset?
plt.hist(dataset['Age'], bins=10, edgecolor='black')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

# Show the plot
plt.show()
```

```python
#2) How many unique values are there for each categorical variable like 'BusinessTravel', 'Department', 'EducationField', etc.?
#1)Business Travel
print("Business Travel Categorical Value")
print(dataset['BusinessTravel'].unique())
print("\n")

print("Department")
print(dataset['Department'].unique())
print("\n")

print("EducationField")
print(dataset['EducationField'].unique())
print("\n")

print("Gender")
print(dataset['Gender'].unique())
print("\n")

print("Job Role")
print(dataset['JobRole'].unique())
print("\n")

print("MaritalStatus")
print(dataset['MaritalStatus'].unique())
print("\n")

print("Over18")
print(dataset['Over18'].unique())
print("\n")

print('OverTime')
print(dataset['OverTime'].unique())
print("\n")
```

```python
#Job and Satisfaction:
#What is the distribution of job roles in the company?
job = dataset['JobRole'].value_counts()
jobDF = pd.DataFrame(job)
jobDF
```

```python
#How satisfied are employees with their jobs and relationships?
job_satisfaction_by_role = dataset.groupby('JobRole')['JobSatisfaction'].mean()

# Create a bar plot
plt.figure(figsize=(10, 6))
job_satisfaction_by_role.plot(kind='bar', color='skyblue')

# Title and labels
plt.title('Average Job Satisfaction by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Average Job Satisfaction')

# Rotate x-labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Work Environment
relationship_satisfaction_by_marital_status = dataset.groupby('MaritalStatus')['RelationshipSatisfaction'].mean()

# Create a bar plot
plt.figure(figsize=(7, 6))
relationship_satisfaction_by_marital_status.plot(kind='bar', color='skyblue')

# Title and labels
plt.title('Average Job Satisfaction by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Average Job Satisfaction')

# Rotate x-labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Is there a correlation between job satisfaction and performance rating?
corr_res = dataset['JobSatisfaction'].corr(dataset['PerformanceRating'])
corr_res
```

```python
#How does attrition vary by gender and marital status?
attritiongender = dataset.groupby('Gender')['Attrition'].value_counts()
attritiongenderDF = pd.DataFrame(attritiongender)
attritiongenderDF
```

```python
attritiongender = dataset.groupby(['Gender', 'Attrition']).size().unstack().fillna(0)

# Create a side-by-side bar plot
ax = attritiongender.plot(kind='bar', figsize=(8, 6))

# Title and labels
plt.title('Attrition by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

# Add labels to the bars
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Are certain departments experiencing higher attrition rates than others?
attritiondept = dataset.groupby('Department')['Attrition'].value_counts()
attritiondeptDF = pd.DataFrame(attritiondept)
attritiondeptDF
```

```python
attritiondept = dataset.groupby(['Department', 'Attrition']).size().unstack().fillna(0)

# Create a side-by-side bar plot
ax = attritiondept.plot(kind='bar', figsize=(10, 6))

# Title and labels
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Count')
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Does attrition vary based on work-life balance scores?
attritionwork = dataset.groupby('WorkLifeBalance')['Attrition'].value_counts()
attritionworkDF = pd.DataFrame(attritionwork)
attritionworkDF
```

```python
attrition_worklife = dataset.groupby(['WorkLifeBalance', 'Attrition']).size().unstack().fillna(0)

# Create a side-by-side bar plot
ax = attrition_worklife.plot(kind='bar', figsize=(10, 6), color=['skyblue', 'lightcoral'])

# Title and labels
plt.title('Attrition by Work Life Balance')
plt.xlabel('Work Life Balance')
plt.ylabel('Count')

# Add legend
plt.legend(title='Attrition', loc='upper right', labels=['No', 'Yes'])

# Add labels to the bars
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Are employees with certain education fields more likely to leave the company?
attritioneducation = dataset.groupby('Education')['Attrition'].value_counts()
attritioneducationDF = pd.DataFrame(attritioneducation)
attritioneducationDF
```

```python
attrition_education = dataset.groupby(['Education', 'Attrition']).size().unstack().fillna(0)

# Create a side-by-side bar plot
ax = attrition_education.plot(kind='bar', figsize=(10, 6), color=['skyblue', 'lightcoral'])

# Title and labels
plt.title('Attrition by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')

# Add legend
plt.legend(title='Attrition', loc='upper right', labels=['No', 'Yes'])

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')
# Show the plot
plt.tight_layout()
plt.show()
```
